## 3. 데이터 전처리

타율에 대한 통계량이 없는 이유는 앞서 타율이 수치형이 아니라 object로 형성 되었기 때문에 describe()함수를 이용해도 나오지 않는다는 것을 확인하여 바꿔주었습니다.

````python
#다음은 '-'를 NaN으로 대체하는 코드입니다. -> '-'이 있는 이유는 선수 등록은 되어 있지만 타석에 들어선적이 없는 선수들이 있기 때문임
kbo_batter_2021.replace('-', np.nan, inplace=True)
#타율이 object이기때문에 다른 수치형인 float으로 바꿔준다.
kbo_batter_2021['AVG'] = kbo_batter_2021['AVG'].astype(float)
````

````python
# 타자의 중요한 지표인 OPS(장타율 + 출루율) 데이터를 받아서 csv파일로 저장
kbo_batter2_2021 = pd.read_csv('/content/drive/MyDrive/2021년 타자 데이터 2.csv')
kbo_batter2_2021
````

````python
#앞선 데이터 프레임을 합치기 위해서는
#두 데이터프레임을 합치려면 공통된 열을 기준으로 합쳐야 합니다. 공통된 열로는 선수명이 있으므로, merge() 함수를 사용하여 두 데이터프레임을 합쳐줄 수 있습니다.
kbo_merged_df = pd.merge(kbo_batter_2021, kbo_batter2_2021, on='선수명', how='inner')
kbo_merged_df
````

````python
import numpy as np
# 먼저 OPS 전체 평균을 구하기
OPS_mean = np.mean(kbo_merged_df['OPS'])
OPS_mean
````

````python
team_ops = kbo_merged_df.groupby('팀명_x')['OPS'].mean().sort_values(ascending=False)
team_ops_df = team_ops.to_frame().reset_index()
team_ops_df
````

````python
# 2021년 KBO 구단별 OPS 구하기
# 보면 우승했던 KT가 역시나 2번째로 높은 OPS를 가지고 있다. 그에 반해 최하위였던 한화는 평균 타자 OPS보다 한참 밑도는 타자 성적표를 가졌다는 것을 확인할 수 있다.
# 여기서 특이점은 SSG구단의 OPS가 우승한 KT보다 더 높은 OPS를 가지고 있다는 점이다. 이를 통해서 SSG구단은 타자 성적은 좋았지만, 다른 부분에서 성적이 저조했다는 것을 간접적으로
# 알 수 있는 대목이었다.
plt.figure(figsize=(12, 6))
plt.plot(team_ops_df['팀명_x'], team_ops_df['OPS'], 'o-', label='구단별 평균 OPS')
plt.axhline(OPS_mean, color='gray', label='전체 타자 평균 OPS')
plt.title('2021 KBO 구단별 OPS')
plt.xlabel('구단')
plt.ylabel('구단별 평균 OPS')
plt.xticks(rotation=45)
plt.legend()
plt.show()
````

````python
import matplotlib.pyplot as plt

# 각 구단별 OPS 평균을 구합니다.
team_ops = kbo_merged_df[kbo_merged_df['G'] >= 100].groupby('팀명_x')['OPS'].mean()

# 산포도를 그립니다.
plt.figure(figsize=(12, 6))
plt.scatter(kbo_merged_df[kbo_merged_df['G'] >= 100]['팀명_x'], kbo_merged_df[kbo_merged_df['G'] >= 100]['OPS'])
for i, txt in enumerate(kbo_merged_df[kbo_merged_df['G'] >= 100]['선수명']):
    plt.annotate(txt, (kbo_merged_df[kbo_merged_df['G'] >= 100]['팀명_x'].iloc[i], kbo_merged_df[kbo_merged_df['G'] >= 100]['OPS'].iloc[i]))
plt.axhline(team_ops.mean(), color='gray', linestyle='-', label='리그 평균')
for team in team_ops.index:
    plt.axvline(x=team, color='gray', linestyle='--')
plt.xticks(rotation=45)
plt.xlabel('구단')
plt.ylabel('OPS')
plt.title('각 구단별 OPS 산포도')
plt.legend()
plt.show()
````

````python
# 경기를 100경기 치룬 타자들만 따로 추려서 구단별 OPS 산포도를 그려보았습니다.

import matplotlib.pyplot as plt

# 각 구단별 OPS 평균을 구합니다.
team_ops = kbo_merged_df[kbo_merged_df['G'] >= 100].groupby('팀명_x')['OPS'].mean()

# 타자별 OPS값에 따라 다른 색상을 지정합니다.
colors = kbo_merged_df[kbo_merged_df['G'] >= 100]['OPS']
cmap = plt.cm.get_cmap('coolwarm')
normalize = plt.Normalize(vmin=colors.min(), vmax=colors.max())
colors = [cmap(normalize(value)) for value in colors]

# 산포도를 그립니다.
plt.figure(figsize=(12, 6))
plt.scatter(kbo_merged_df[kbo_merged_df['G'] >= 100]['팀명_x'], kbo_merged_df[kbo_merged_df['G'] >= 100]['OPS'], c=colors)
plt.axhline(team_ops.mean(), color='gray', linestyle='-', label='리그 평균')
for team in team_ops.index:
    plt.axvline(x=team, color='gray', linestyle='--')
plt.xticks(rotation=45)
plt.xlabel('구단')
plt.ylabel('OPS')
plt.title('각 구단별 OPS 산포도')
plt.legend()
plt.show()
````

### 두 팀의 투수들의 데이터를 따로 프레임으로 만들어주었습니다.

````python
team_pitcher_records = kbo_pitcher_2021.groupby('팀명')
````

````python
kt_pitchers = team_pitcher_records.get_group('KT')
ssg_pitchers = team_pitcher_records.get_group('SSG')
````

````python
kt_pitchers
````

````python
ssg_pitchers
````

````python
# KT와 SSG 투수들의 WHIP 데이터 가져오기
kt_pitchers_whip = kt_pitchers['WHIP']
ssg_pitchers_whip = ssg_pitchers['WHIP']

# KT와 SSG 팀의 평균 WHIP 계산
kt_avg_whip = kt_pitchers_whip.mean()
ssg_avg_whip = ssg_pitchers_whip.mean()

# KBO 리그 평균 WHIP 계산
kbo_avg_whip = kbo_pitcher_2021['WHIP'].mean()

# 그래프 그리기
plt.figure(figsize=(10, 6))
plt.plot(kt_pitchers_whip, 'o-', label='KT 투수들 WHIP')
plt.plot(ssg_pitchers_whip, 'o-', label='SSG 투수들 WHIP')
plt.axhline(kt_avg_whip, color='red', linestyle='--', label='KT 평균 WHIP')
plt.axhline(ssg_avg_whip, color='blue', linestyle='--', label='SSG 평균 WHIP')
plt.axhline(kbo_avg_whip, color='gray', linestyle='--', label='리그 평균 WHIP')
plt.title('KT vs. SSG 투수들의 WHIP 비교')
plt.xlabel('선발 투수')
plt.ylabel('WHIP')
plt.legend()
plt.xticks(rotation=45)
plt.show()
````

결론 : kt와 ssg 두 팀 차이는 투수에서 나온다는 것을 알 수 있었다. ssg투수들의 whip는 리그 평균 수준이고 kt 투수들의 whip 간의 차이가 꽤 큰것을 시각화를 통해서 알 수 있었다.
